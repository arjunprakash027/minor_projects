```python
import os
import cv2
path = r"D:\something to do with studies\datas\face data\Humans\1 (14).jpg"
path1 = cv2.imread(path)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(path1)
```

```python
import imgaug as ia
import imgaug.augmenters as iaa
from imgaug.augmentables.bbs import BoundingBox, BoundingBoxesOnImage
```

```python
bbs = BoundingBoxesOnImage([
    BoundingBox(x1=200, y1=200, x2=550, y2=320),
    BoundingBox(x1=400, y1=400, x2=500, y2=500),
    BoundingBox(x1=220, y1=400, x2=300, y2=500)],
    shape = path1.shape)
```

```python
seq = iaa.Sequential([
    iaa.AdditiveGaussianNoise(scale=0.05*255),
    iaa.Affine(translate_px={"x": (1, 5)})
])
```

```python
images_aug, bbs_aug = seq(images=path1, bounding_boxes=bbs)
```

```python
plt.imshow(bbs.draw_on_image(path1,size=10))
```
