```python
cols=list(set(df.columns)-set(df.describe().columns))
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for eachcol in cols:
  le=LabelEncoder()
  df[eachcol]=le.fit_transform(df[eachcol])
```

```python
df
```

```python
df['class'].head(10)
```

```python
x=df[['intercolumnar distance ', 'upper margin', 'lower margin',
       'exploitation', 'row number', 'modular ratio ', 'interlinear spacing ',
       'weight ', 'peak number ', 'modular ratio']].values
```

```python
y=df['class'].values
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(xtrain,ytrain.ravel())
```

```python
ypred=knn.predict(xtest)


```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypred)
```
